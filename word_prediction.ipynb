```python
import pandas as pd
import string
import numpy as np
import matplotlib.pyplot as mp 
```

```python
t=pd.read_csv("TestReviews.csv")
```

```python
t=t["review"]
```

```python
t.head()
```

```python
t.shape
```

```python
t=t.sample(20)
```

```python
t=list(t)
```

```python
t=[i.replace("\n","").replace("\t","") for i in t]
```

```python
punch=string.punctuation
```

```python
tt=[]
for i in t:
    tt.append("".join([char for char in i if char not in punch]))
```

```python
t
```

```python
tt
```

```python
k=""
ttt=[]
for i in range(len(tt)):
    k=k+tt[i]
ttt.append(k)
```

```python
ttt
```

```python
Data=pd.DataFrame(ttt[0].split())
```

```python
Data
```

```python
def words_break(data):
    kk=[]
    for i in range(1,len(data)):
        kk.append([data[:i],data])
    return kk
ko=Data[0].apply(words_break)
```

```python
kk=[]
for i in ko:
    for j in i:
        kk.append(j)
```

```python
word=[]
words=[]
for i,j in kk:
    word.append(i)
    words.append(j)
```

```python
word
```

```python
words
```

```python
d={"word":word,"words":words}
```

```python
Train=pd.DataFrame(d)
```

```python
Train.head()
```

```python
Train=Train.sample(frac=1).reset_index(drop=True)
```

```python
Train.shape
```

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from keras import Sequential
from keras.layers import *
from keras.models import Model
from tensorflow.keras.callbacks import EarlyStopping as ES
```

```python
tok=tf.keras.preprocessing.text.Tokenizer(
    num_words=None,
    filters='!"#$%&()*+,-./:;<=>?@[\\]^_`{|}~\t\n',
    lower=True,
    split=' ',
    char_level=False,
)
```

```python
tok1=tf.keras.preprocessing.text.Tokenizer(
    num_words=None,
    filters='!"#$%&()*+,-./:;<=>?@[\\]^_`{|}~\t\n',
    lower=True,
    split=' ',
    char_level=False,
)
```

```python
tok.fit_on_texts(Train["word"])
```

```python
tok1.fit_on_texts(Train["words"])
```

```python
tok.word_index
```

```python
tok1.word_index
```

```python
x=np.array(tok.texts_to_sequences(Train["word"]))
x
```

```python
y=np.array(tok1.texts_to_sequences(Train["words"]))
y
```

```python
i=Input(shape=(1,))
e=Embedding(len(tok.word_index)+1,output_dim=20)(i)
r=SimpleRNN(2000)(e)
d1=Dense(2000,activation="selu")(r)
d2=Dense(1000,activation="selu")(d1)
d3=Dense(500,activation="selu")(d2)
d4=Dense(350,activation="selu")(d3)
o=Dense(len(tok1.word_index)+1,activation="softmax")(d4)
```

```python
model=Model(inputs=i,outputs=o)
```

```python
model.summary()
```

```python
es=ES(monitor="val_loss",
    min_delta=0.01,
    patience=4,
    verbose=1,
    mode="auto",
    baseline=None,
    restore_best_weights=True,
    start_from_epoch=0,)
```

```python
model.compile(loss="sparse_categorical_crossentropy",optimizer="adam",metrics=["accuracy"])
```

```python
history=model.fit(x,y,validation_split=0.2,epochs=5,callbacks=es)
```

```python
mp.plot(history.history["loss"],label="loss")
mp.plot(history.history["val_loss"],label="val_loss")
mp.legend();
```

```python
mp.plot(history.history["accuracy"],label="accuracy")
mp.plot(history.history["val_accuracy"],label="val_accuracy")
mp.legend();
```
